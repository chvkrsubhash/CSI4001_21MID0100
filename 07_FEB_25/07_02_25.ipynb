```python
def gender_feature(word):
    return {'last_letter':word[-1]}
```

```python
gender_feature('kriti')
```

```python
import nltk
from nltk.corpus import names
```

```python
nltk.download('names')
```

```python
names.words()
```

```python
print(len(names.words()))
```

```python
labeled_names = ([(names,'male') for name in names.words('male.txt')]+[(names,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
featuresets = [(gender_feature(str(n)), gender) for (n, gender) in labeled_names]
```

```python
train_set,test_set = featuresets[5000:],featuresets[:2000]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_feature('Subhash'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
